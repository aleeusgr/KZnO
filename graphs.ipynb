```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()
```

### Visualisation

```python
print('Simple')
# nx.draw_circular(G, with_labels=True)
nx.draw(G,with_labels = True)
plt.show()
```

```python
print('Advanced')

# Some matplotlib options
plt.figure(figsize=(8,8))
plt.axis('off')

# generate the layout and place nodes and edges
layout = nx.circular_layout(G)

# plot nodes, labels, and edges with options
nx.draw_networkx_nodes(G, pos=layout, node_size=500, alpha=0.8)
nx.draw_networkx_edges(G, pos=layout, width=3, style='dotted',
                       edge_color='orange')
nx.draw_networkx_labels(G, pos=layout, font_size=15)

plt.show()
```

```python
print('centrality')
G.degree(32)
sorted(G.degree(), key=lambda x:x[1], reverse=True)[:5] #5 highest.


degree_centrality = nx.degree_centrality(G)
# Number of edges in the node. Not much context withing the whole network
closeness_centrality = nx.closeness_centrality(G)
# average distance to all the nodes, theoretically message passed from highest-closeness, will spread in the network most quickly 
eigenvector_centrality = nx.eigenvector_centrality_numpy(G)
# well connected, and its neighbours are well-connected.
betweeness_centrality = nx.betweenness_centrality(G)
# chokepoints between clusters
nx.set_node_attributes(G, degree_centrality, 'degree centrality')
```
